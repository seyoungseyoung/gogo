```python
import pandas as pd
import numpy as np
import ast

df_ids = pd.read_csv('/Users/jaewooshin/Desktop/Everything/Uni/2025-1/주식 프로젝트/대충 데이터/fill_records_auto_128_links.csv')
df_sorted_list = pd.read_csv('/Users/jaewooshin/Downloads/transformer_all_results (2).csv')
df_links_2019 = pd.read_csv('/Users/jaewooshin/Desktop/Everything/Uni/2025-1/주식 프로젝트/대충 데이터/SamsungE_2019_processed.csv')
df_links_2020 = pd.read_csv('/Users/jaewooshin/Desktop/Everything/Uni/2025-1/주식 프로젝트/대충 데이터/SamsungE_2020_processed.csv')
df_links_2021 = pd.read_csv('/Users/jaewooshin/Desktop/Everything/Uni/2025-1/주식 프로젝트/대충 데이터/SamsungE_2021_processed.csv')
df_links_2022 = pd.read_csv('/Users/jaewooshin/Desktop/Everything/Uni/2025-1/주식 프로젝트/대충 데이터/SamsungE_2022_processed.csv')
df_links_2023 = pd.read_csv('/Users/jaewooshin/Desktop/Everything/Uni/2025-1/주식 프로젝트/대충 데이터/SamsungE_2023_processed.csv')
df_links_2024 = pd.read_csv('/Users/jaewooshin/Desktop/Everything/Uni/2025-1/주식 프로젝트/대충 데이터/SamsungE_2024_processed.csv')
df_links_2025 = pd.read_csv('/Users/jaewooshin/Desktop/Everything/Uni/2025-1/주식 프로젝트/대충 데이터/SamsungE_2025_processed.csv')

dfs = [
    df_links_2019,
    df_links_2020,
    df_links_2021,
    df_links_2022,
    df_links_2023,
    df_links_2024,
    df_links_2025
]

df_all_links = pd.concat(dfs, ignore_index=True)
```

```python
df_sorted_list['Refined_Date'] = [row[0:4] + row[5:7] + row[8:10] for row in df_sorted_list['Date']]
sorted_dataframe = df_sorted_list[['Refined_Date', 'Sorted_List']]
ids_dataframe = df_ids[['Date', 'IDs']]
links_dataframe = df_all_links[['Date', 'Title', 'Link', 'ID']]
```

```python
rearranged_ids_full_lst = []
rearranged_links_full_lst_1 = []
rearranged_links_full_lst_2 = []
rearranged_links_full_lst_3 = []

rearranged_links_full_lst_4 = []
rearranged_links_full_lst_5 = []
rearranged_links_full_lst_6 = []
rearranged_links_full_lst_7 = []
rearranged_links_full_lst_8 = []
rearranged_links_full_lst_9 = []
rearranged_links_full_lst_10 = []



rearranged_titles_full_lst_1 = []
rearranged_titles_full_lst_2 = []
rearranged_titles_full_lst_3 = []

rearranged_titles_full_lst_4 = []
rearranged_titles_full_lst_5 = []
rearranged_titles_full_lst_6 = []
rearranged_titles_full_lst_7 = []
rearranged_titles_full_lst_8 = []
rearranged_titles_full_lst_9 = []
rearranged_titles_full_lst_10 = []




for i in range(len(sorted_dataframe)):
    row = sorted_dataframe.iloc[i]
    date = np.int64(row['Refined_Date'])
    
    row_ids = ids_dataframe[ids_dataframe['Date'] == date]
    row['Sorted_List'] = ast.literal_eval(row['Sorted_List'])
    
    lst_ids = ast.literal_eval(row_ids['IDs'].squeeze())
    rearranged_ids = [lst_ids[i-1] for i in row['Sorted_List']]
    
    links_list_i = []
    titles_list_i = []
    for id in rearranged_ids:
        correct_link_series = links_dataframe[links_dataframe['ID'] == id]['Link']
        links_list_i.append(correct_link_series.squeeze())
        correct_title_series = links_dataframe[links_dataframe['ID'] == id]['Title']
        titles_list_i.append(correct_title_series.squeeze())

    rearranged_ids_full_lst.append(rearranged_ids)

    rearranged_links_full_lst_1.append(links_list_i[0])
    rearranged_links_full_lst_2.append(links_list_i[1])
    rearranged_links_full_lst_3.append(links_list_i[2])
    
    rearranged_links_full_lst_4.append(links_list_i[3])
    rearranged_links_full_lst_5.append(links_list_i[4])
    rearranged_links_full_lst_6.append(links_list_i[5])
    rearranged_links_full_lst_7.append(links_list_i[6])
    rearranged_links_full_lst_8.append(links_list_i[7])
    rearranged_links_full_lst_9.append(links_list_i[8])
    rearranged_links_full_lst_10.append(links_list_i[9])


    rearranged_titles_full_lst_1.append(titles_list_i[0])
    rearranged_titles_full_lst_2.append(titles_list_i[1])
    rearranged_titles_full_lst_3.append(titles_list_i[2])

    rearranged_titles_full_lst_4.append(titles_list_i[3])
    rearranged_titles_full_lst_5.append(titles_list_i[4])
    rearranged_titles_full_lst_6.append(titles_list_i[5])
    rearranged_titles_full_lst_7.append(titles_list_i[6])
    rearranged_titles_full_lst_8.append(titles_list_i[7])
    rearranged_titles_full_lst_9.append(titles_list_i[8])
    rearranged_titles_full_lst_10.append(titles_list_i[9])
```

```python
df_sorted_list['Sorted IDs'] = rearranged_ids_full_lst
df_sorted_list['Sorted Title 1'] = rearranged_titles_full_lst_1
df_sorted_list['Sorted Links 1'] = rearranged_links_full_lst_1
df_sorted_list['Sorted Title 2'] = rearranged_titles_full_lst_2
df_sorted_list['Sorted Links 2'] = rearranged_links_full_lst_2
df_sorted_list['Sorted Title 3'] = rearranged_titles_full_lst_3
df_sorted_list['Sorted Links 3'] = rearranged_links_full_lst_3

df_sorted_list['Sorted Title 4'] = rearranged_titles_full_lst_4
df_sorted_list['Sorted Links 4'] = rearranged_links_full_lst_4
df_sorted_list['Sorted Title 5'] = rearranged_titles_full_lst_5
df_sorted_list['Sorted Links 5'] = rearranged_links_full_lst_5
df_sorted_list['Sorted Title 6'] = rearranged_titles_full_lst_6
df_sorted_list['Sorted Links 6'] = rearranged_links_full_lst_6
df_sorted_list['Sorted Title 7'] = rearranged_titles_full_lst_7
df_sorted_list['Sorted Links 7'] = rearranged_links_full_lst_7
df_sorted_list['Sorted Title 8'] = rearranged_titles_full_lst_8
df_sorted_list['Sorted Links 8'] = rearranged_links_full_lst_8
df_sorted_list['Sorted Title 9'] = rearranged_titles_full_lst_9
df_sorted_list['Sorted Links 9'] = rearranged_links_full_lst_9
df_sorted_list['Sorted Title 10'] = rearranged_titles_full_lst_10
df_sorted_list['Sorted Links 10'] = rearranged_links_full_lst_10
```
